# **Projeto: Previsão de Intenção de Compra de Clientes em Loja Web**

**Descrição do Projeto**

Neste projeto, nosso objetivo é criar um sistema inteligente para antecipar a intenção de compra dos clientes em um site de e-commerce. Imagine poder prever quais clientes têm maior probabilidade de realizar compras online, com base em suas características e comportamentos passados. Essa capacidade de prever a intenção de compra não só aprimorará a experiência do cliente, mas também permitirá que a empresa direcione seus esforços de marketing de forma mais eficaz.

**Objetivo**

Queremos desenvolver um modelo preditivo capaz de analisar os padrões de comportamento dos clientes e identificar sinais que indicam a propensão deles para realizar compras no site da empresa. Para isso, vamos usar uma base de dados que contém informações detalhadas sobre os clientes, incluindo:

Dados demográficos (idade, renda, etc.)

Informações sobre compras anteriores



# Base de dados:

Year_Birth: Ano de nascimento do cliente.

Education: Nível de escolaridade do cliente.

Marital_Status: Estado civil do cliente.

Income: Renda anual da família do cliente.

Kidhome: Número de crianças na casa do cliente.

Recency: Número de dias desde a última compra do cliente.

Complain: 1 se o cliente reclamou nos últimos 2 anos, 0 caso contrário.

MntWines: Valor gasto em vinhos nos últimos 2 anos.

MntFruits: Valor gasto em frutas nos últimos 2 anos.

MntMeatProducts: Valor gasto em carnes nos últimos 2 anos.

MntFishProducts: Valor gasto em peixes nos últimos 2 anos.

MntSweetProducts: Valor gasto em doces nos últimos 2 anos.

MntGoldProds: Valor gasto em produtos de ouro nos últimos 2 anos.

NumDealsPurchases: Número de compras feitas com desconto

NumStorePurchases: Número de compras feitas diretamente nas lojas.

NumWebVisitsMonth: Número de visitas ao site da empresa no último mês.






**WebPurchases: Número de compras feitas pelo site da empresa.**

**Preparação dos Dados**

**Exploração e Limpeza:** Analisar e limpar os dados para garantir que estejam prontos para a modelagem.

```python
# Biblioteca de dataframe
import pandas as pd
import numpy as np

# Biblioteca para visualização de dados e análise gráfica
import plotly_express as px

# Bibliotecas de preparação do modelo
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

# Biblioteca do modelo
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# Biblioteca de avaliação do modelo
from sklearn.metrics import classification_report, confusion_matrix

# Bibliotecas de codificação e padronização
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from scipy.stats import zscore
from scipy.stats import randint
from scipy.stats import loguniform
from sklearn.model_selection import RandomizedSearchCV
```

```python
base = pd.read_csv('marketing_campaign.csv', delimiter=';')
```

```python
# Visualização do dataframe
base.head()
```

```python
# Análise de tipos de dados e possíveis dados faltantes
base.info()
```

Há duas colunas de dados escritos, serão transformadas em códigos numéricos.

A coluna de salário tem 24 dados nulos ou faltantes.

```python
# Usando describe para observar a presença de outliers
base.describe()
```

```python
'''
Como temos colunas de caracteres vamos transformá-los em numéricos primeiro
'''
encoder = LabelEncoder()
base['Education'] = encoder.fit_transform(base['Education'])
base['Marital_Status'] = encoder.fit_transform(base['Marital_Status'])
```

```python
base.head()
```

```python
'''
Após observar a presença de possíveis outliers vamos usar z-score para limpar os dados
'''
base = base.astype(float)
# Achando o z-score para cada variável numérica do dataframe
z_score = base.apply(zscore)
mask = ~(z_score.abs() > 3).any(axis=1)
base_limpa = base[mask]
```

```python
# Observando quantos dados foram eliminados e a possível presença de outliers
base_limpa.describe()
```

```python
'''
Tendo em vista que já foi feito a limpeza de outliers e codificação de caracteres é razoável eliminar os dados nulos da coluna 'Income', por causa de seu pequeno volume.
'''
base_limpa= base_limpa.dropna(subset=['Income'])
base_limpa = base_limpa[base_limpa['Income'] != 666666]
```

# ETAPA 2:
**Pré-processamento**

**Análise Correlação:** Verifique a correlação entre as váriaveis e análise se há espaço para retirar váriaveis que não te parecem importantes.

**Codificação de Variáveis Categóricas:** Transformar variáveis categóricas em um formato que os modelos de machine learning possam interpretar.


**Separe a base em Y, X e Treino e teste:**: Faça a separação da base.

**Realize a padronização dos dados**: Padronize os dados para garantir eficiência no modelo e eficácia.








```python
# Gráfico de correlação entre as variáveis
corr =  base_limpa.corr()
fig = px.imshow(
    corr,
    text_auto=True,
    aspect='full',
    color_continuous_scale='Tealgrn',
    template='seaborn',
    title='Matriz de correlação'
)
fig.show()
```

```python
'''
Como a codificação de variáveis categóricas foi feita na etapa acima, não é necessário realizá-la de novo.
Nessa etapa, a coluna 'Complain' será eliminada porque os valores relativos à presença de reclamações foram eliminados por conter outliers em alguma outra variável.
Além disso, os valores serão padronizados e o PCA será utilizado para selecionar as variáveis que podem explicar melhor a variável target
'''

base_limpa = base_limpa.drop(columns='Complain')

# Separando em X e Y.
X = base_limpa.drop(columns='WebPurchases')
Y = base_limpa['WebPurchases']
```

```python
# Padronizando valores das variáveis
pad_scaler = StandardScaler()
X_pad = pad_scaler.fit_transform(X)
X_pad = pd.DataFrame(X_pad, columns=X.columns, index=X.index)
```

```python
X_pad.head()
```

```python
# Utilizando o PCA para observar as variáveis necessárias ao modelo
pca = PCA(n_components=10)
pca.fit(X_pad)
```

```python
# Gráfico de variância explicada acumulada
exp_cul = np.cumsum(pca.explained_variance_ratio_)

fig = px.area(
    x=range(1, exp_cul.shape[0]+1),
    y=exp_cul,
    labels={'x':'Número de componentes', 'y':'Variância explicada'},
    template = 'seaborn',
    title='Gráfico de PCA'
)
fig.update_traces(line_color='#2aa889')
fig.show()
```

Com 10 componentes/variáveis pode-se explicar 91% da variabilidade dos dados da base.

```python
X_pca = pca.fit_transform(X_pad)
```

```python
# Dividindo a base em treino e teste
X_train, X_test, Y_train, Y_test = train_test_split(X_pca, Y, test_size=0.3, random_state=42)
```

# ETAPA 3:

**Modelagem**

Escolha ao menos 2 técnicas de machine learning e rode 2 modelos, afim de identificar qual tem o melhor resultado para essa base. Lembrando que estamos lidando com uma classificação binária.

```python
# Primeiro modelo treinado será o de regressão logística
lr_model = LogisticRegression(random_state=42)

# Procurando melhores hiperparâmetros para o modelo
log_par = {
    'C': loguniform(1e-3, 1e3),
    'penalty':[None, 'l2'],
    'solver': ['newton-cg', 'lbfgs']
}
rsearch_log = RandomizedSearchCV(
    estimator=lr_model,
    param_distributions=log_par,
    n_iter=100,
    n_jobs=-1,
    cv=5
)
rsearch_log.fit(X_train, Y_train)

```

```python
# Pontuação de treino do modelo de regressão logística
melhor_logreg = rsearch_log.best_score_
print(f'Melhor pontuação do modelo: {melhor_logreg:.2f}')
```

```python
# Treino de modelo de random forest
rf_model = RandomForestClassifier(random_state=42)

# Seleção de hiperparâmetros
rf_par = {
    'n_estimators':randint(50, 200),
    'criterion':['gini', 'entropy'],
    'max_depth':range(1,30),
    'min_samples_split':range(2,30),
    'max_features':['sqrt', 'log2'],
}

# Configurando o modelo com hiperparâmetros
rsearch_rf = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=rf_par,
    cv = 5,
    n_iter=100,
    n_jobs=-1,
)
rsearch_rf.fit(X_train, Y_train)
```

```python
# Melhor pontuação de treino do random forest
melhor_rf = rsearch_rf.best_score_
print(f"Melhor pontuação do modelo: {melhor_rf:.2f}")
```

A pontuação de treino do modelo de random forest foi ligeiramente melhor, mas a pontuação dos dois modelos foi satisfatória.

# ETAPA 4:

**Avaliação**

Avalie os resultados encontrados nos dois modelos e identifique qual te pareceu realizar melhor as previsões.

Utilize além das métricas padrões a matriz de confusão.

```python
# Teste do modelo de regressão logística
Y_pred_lr = rsearch_log.predict(X_test)

# Gráfico da matriz de confusão
conf_log = confusion_matrix(Y_test, Y_pred_lr)

classes = ['Não compra web', 'Compra web']
fig = px.imshow(
    conf_log,
    text_auto=True,
    aspect='auto',
    template='seaborn',
    x=classes,
    y=classes,
    labels=dict(x='Predição', y='Real', color='Contagem'),
    color_continuous_scale='Temps',
    title='Matriz de confusão - Regressão logística'
)
fig.show()
```

A Matriz de confusão do modelo de regressão logística mostrou uma boa quantidade de acertos no teste.

```python
# Relatório de classificação do modelo de regressão logística
print("Relatório de classificação")
print(classification_report(Y_test, Y_pred_lr))
```

O modelo demonstrou uma acurácia e uma média ponderada de 0.83, ou seja, a diferença entre treino e teste foi ínfima. Isso demonstra que o modelo foi bem preparado e treinado.

```python
# Teste de modelo de random forest
Y_pred_rf = rsearch_rf.predict(X_test)

# Gráfico de matriz de confusão
conf_rf = confusion_matrix(Y_test, Y_pred_rf)

classes = ['Não compra web', 'Compra web']
fig = px.imshow(
    conf_rf,
    text_auto=True,
    aspect='auto',
    template='seaborn',
    x=classes,
    y=classes,
    labels=dict(x='Predição', y='Real', color='Contagem'),
    color_continuous_scale='Temps',
    title='Matriz de confusão - Random forest'
)
fig.show()
```

Assim como na etapa de treino, o modelo de random forest foi ligeiramente melhor que o de regressão logística na fase de teste.

```python
# Relatório de classificação do modelo de random forest
print("Relatório de classificação")
print(classification_report(Y_test, Y_pred_rf))
```

O modelo de random forest foi superior em todas as métricas, incluindo a mais importante, que era conseguir prever clientes que comprariam pelo site da empresa, com um recall de 0.92 e uma diferença de 42 verdadeiros positivos, em relação ao modelo de regressão logística.

Portanto, para essa variável, e essa base de dados, em questão, o modelo de random forest parece ser o mais indicado.
